# Dataset Preparation

```python
import pandas as pd
```

## human-AI-generatedTextCorpus

```python
# wiki or news
domain = "wiki"

human_features_df =  pd.read_pickle("../raw_data/en_wiki_human_features_df.pkl")
gpt_features_df =  pd.read_pickle("../raw_data/en_wiki_gpt_features_df.pkl")
```

```python
human_features_df["Label"] = "Human"
human_features_df.rename(columns={"text":"Text"}, inplace=True)
human_features_df = human_features_df[["Label", "Text"]]
human_features_df.head()
```

```python
gpt_features_df["Label"] = "OpenAI"
gpt_features_df.rename(columns={"text":"Text"}, inplace=True)
gpt_features_df = gpt_features_df[["Label", "Text"]]
gpt_features_df.head()
```

```python
df_1 = pd.concat([human_features_df, gpt_features_df])
df_1["word_count"] = df_1.apply(lambda row: len(row["Text"].split()), axis=1)
df_1.reset_index(drop=True, inplace=True)
df_1
```

## GPT vs. Human: A Corpus of Research Abstracts

```python
df = pd.read_csv('../raw_data/gptvshuman_dataset.csv')
df
```

```python
df["Label"] = df.apply(lambda row: "OpenAI" if row["ai_generated"] else "Human", axis = 1)
df
```

```python
df.rename(columns={"abstract":"Text"}, inplace=True)
df
```

```python
df_2 = df[["Label", "Text"]]
df_2["word_count"] = df_2.apply(lambda row: len(row["Text"].split()), axis=1)
df_2.reset_index(drop=True, inplace=True)
df_2
```

## AI vs Human Scientific Text Dataset (AIGTxt)

```python
df = pd.read_csv('../raw_data/AIGTxt.csv', delimiter=";")
df
```

```python
df_human = pd.DataFrame(df["Human-Generated"])
df_human.rename(columns={"Human-Generated":"Text"}, inplace=True)
df_human["Label"] = "Human"
df_human
```

```python
df_ai = pd.DataFrame(df["ChatGPT-Generated"])
df_ai.rename(columns={"ChatGPT-Generated":"Text"}, inplace=True)
df_ai["Label"] = "OpenAI"
df_ai
```

```python
df_3 = pd.concat([df_human, df_ai])
df_3["word_count"] = df_3.apply(lambda row: len(row["Text"].split()), axis=1)
df_3.reset_index(drop=True, inplace=True)
df_3
```

## Human vs. LLM Text Corpus

```python
df = pd.read_csv('../raw_data/humanvsllm_data.csv')
df
```

```python
source_mapping = {
    "Human": "Human",
    "GPT-3.5": "OpenAI",
    "Text-Davinci-003": "OpenAI",
    "Text-Davinci-002": "OpenAI",
    "OPT-1.3B": "Meta",
    "OPT-30B": "Meta",
    "PaLM-2": "Google",
    "LLaMA-30B": "Meta",
    "LLaMA-65B": "Meta",
    "LLaMA-13B": "Meta",
    "LLaMA-7B": "Meta",
    "Flan-T5-Base": "Google",
    "Flan-T5-Large": "Google",
    "Flan-T5-Small": "Google",
    "OPT-2.7B": "Meta",
    "Flan-T5-XXL": "Google",
    "Flan-T5-XL": "Google",
    "GPT-4": "OpenAI",
    "OPT-6.7B": "Meta",
    "OPT-125M": "Meta",
    "OPT-350M": "Meta",
    "OPT-13B": "Meta",
    "GPT-J": "OpenAI",
    "Claude-Instant-v1": "Anthropic",
    "GPT-NeoX": "OpenAI",
    "LLaMA-2-70B": "Meta",
    "Claude-v1": "Anthropic",
    "Gemini-Pro": "Google",
    "LLaMA-2-7B": "Meta"
}
```

```python
df = df[df["source"].isin(source_mapping.keys())][["source", "text", "word_count"]]
df.rename(columns={"source":"Label"}, inplace=True)
df
```

```python
df["Text"] = df["text"].str.replace(r"[\n\r]", " ", regex=True)
```

```python
df.head()
```

```python
df_4 = pd.DataFrame(df[["Text", "word_count"]])
df_4["Label"] = df.apply(lambda row: source_mapping[row["Label"]], axis=1)
df_4.reset_index(drop=True, inplace=True)
df_4
```

## Final Dataset

```python
df_final = pd.concat([df_1, df_2, df_3, df_4])
df_final.shape
```

```python
df_final["basic_text"] = df_final["Text"].str.upper()
df_final.head()
```

```python
df_final = df_final.drop_duplicates(subset=['basic_text'])
df_final.drop("basic_text", axis=1, inplace=True)
df_final.shape
```

```python
df_short = df_final[df_final["word_count"] <= 120]
df_short.reset_index(drop=True, inplace=True)
df_short.Label.value_counts(dropna=False)
```

```python
df_tmp = df_final[df_final["Label"] == "Anthropic"]
df_final_mini = df_short[df_short["Label"] != "Human"].groupby("Label").sample(n=30000, random_state=42)
df_final_mini = pd.concat([df_final_mini, df_tmp, df_short[df_short["Label"] == "Human"]])
df_final_mini.Label.value_counts(dropna=False)
```

```python
df_final_mini.to_csv('../custom_dataset_v2_mini.csv', sep=";", index_label="ID")
```
