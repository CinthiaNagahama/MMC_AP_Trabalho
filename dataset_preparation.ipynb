# Dataset Preparation

```python
import pandas as pd
```

## human-AI-generatedTextCorpus

```python
# wiki or news
domain = "wiki"

human_features_df =  pd.read_pickle("../raw_data/en_wiki_human_features_df.pkl")
gpt_features_df =  pd.read_pickle("../raw_data/en_wiki_gpt_features_df.pkl")
```

```python
human_features_df["source"] = "human"
human_features_df = human_features_df[["source", "text"]]
```

```python
human_features_df.head()
```

```python
gpt_features_df["source"] = "openai"
gpt_features_df = gpt_features_df[["source", "text"]]
```

```python
gpt_features_df.head()
```

```python
df_1 = pd.concat([human_features_df, gpt_features_df])
df_1.reset_index(drop=True, inplace=True)
df_1
```

## GPT vs. Human: A Corpus of Research Abstracts

```python
df = pd.read_csv('../raw_data/gptvshuman_dataset.csv')
df
```

```python
df["source"] = df.apply(lambda row: "openai" if row["ai_generated"] else "human", axis = 1)
df
```

```python
df.rename(columns={"abstract":"text"}, inplace=True)
df
```

```python
df_2 = df[["source", "text"]]
df_2.reset_index(drop=True, inplace=True)
df_2
```

## AI-Generated Text Detection (Multi-Model)

```python
df = pd.read_csv('../raw_data/benchmark_ai_detection_multimodel_2026.csv')
df
```

```python
df["source_model"].value_counts(dropna=False)
```

```python
model_mapping = {
    "Human": "human",
    "Gemini-2_Pro": "google",
    "GPT-5_Preview": "openai",
    "Claude-4_Opus": "anthropic"
}
```

```python
df.rename(columns={"text_content":"text"}, inplace=True)
df["source"] = df.apply(lambda row: model_mapping[row["source_model"]], axis=1)
df
```

```python
df_3 = df[["source", "text"]]
df_3
```

## AI vs Human Scientific Text Dataset (AIGTxt)

```python
df = pd.read_csv('../raw_data/AIGTxt.csv', delimiter=";")
df
```

```python
df_human = pd.DataFrame(df["Human-Generated"])
df_human.rename(columns={"Human-Generated":"text"}, inplace=True)
df_human["source"] = "human"
df_human
```

```python
df_ai = pd.DataFrame(df["ChatGPT-Generated"])
df_ai.rename(columns={"ChatGPT-Generated":"text"}, inplace=True)
df_ai["source"] = "openai"
df_ai
```

```python
df_4 = pd.concat([df_human, df_ai])
df_4.reset_index(drop=True, inplace=True)
df_4
```

## Human vs. LLM Text Corpus

```python
df = pd.read_csv('../raw_data/humanvsllm_data.csv')
df
```

```python
source_mapping = {
    "Human": "human",
    "GPT-3.5": "openai",
    "Text-Davinci-003": "openai",
    "Text-Davinci-002": "openai",
    "OPT-1.3B": "meta",
    "OPT-30B": "meta",
    "PaLM-2": "google",
    "LLaMA-30B": "meta",
    "LLaMA-65B": "meta",
    "LLaMA-13B": "meta",
    "LLaMA-7B": "meta",
    "Flan-T5-Base": "google",
    "Flan-T5-Large": "google",
    "Flan-T5-Small": "google",
    "OPT-2.7B": "meta",
    "Flan-T5-XXL": "google",
    "Flan-T5-XL": "google",
    "GPT-4": "openai",
    "OPT-6.7B": "meta",
    "OPT-125M": "meta",
    "OPT-350M": "meta",
    "OPT-13B": "meta",
    "GPT-J": "openai",
    "Claude-Instant-v1": "anthropic",
    "GPT-NeoX": "openai",
    "LLaMA-2-70B": "meta",
    "Claude-v1": "anthropic",
    "Gemini-Pro": "google",
    "LLaMA-2-7B": "meta"
}
```

```python
df = df[df["source"].isin(source_mapping.keys())][["source", "text"]]
df
```

```python
df_5 = pd.DataFrame(df["text"])
df_5["source"] = df.apply(lambda row: source_mapping[row["source"]], axis=1)
df_5.reset_index(drop=True, inplace=True)
df_5
```

## Final Dataset

```python
df_final = pd.concat([df_1, df_2, df_3, df_4, df_5])
df_final.shape
```

```python
df_final["basic_text"] = df_final["text"].str.upper()
df_final.head()
```

```python
df_final = df_final.drop_duplicates(subset=['basic_text'])
df_final.shape
```

```python
df_final.source.value_counts(dropna=False)
```

```python
# df_final.to_csv('../custom_dataset_v1.csv')
```

```python
df_tmp = df_final[df_final["source"] == "anthropic"]
df_final_mini = df_final[df_final["source"] != "anthropic"].groupby("source").sample(n=50000, random_state=42)
df_final_mini = pd.concat([df_final_mini, df_tmp])
df_final_mini.source.value_counts(dropna=False)
```

```python
df_final_mini.to_csv('../custom_dataset_v1_mini.csv')
```
